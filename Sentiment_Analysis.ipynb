```r
data <- read.csv(url("https://raw.githubusercontent.com/wyshnavighag/Sentiment-analysis/main/apple.csv"),na.strings='..')
str(data)
```

```r
#build corpus
install.packages('tm',dependencies = TRUE)
library(tm)
corpus <- iconv(data$text, to= "utf-8-mac")
corpus <- Corpus(VectorSource(corpus))
inspect(corpus[1:5])
```
